## ðŸ§ª **Lesson 2: Create a Simple Learning Component**

Let's build a mini version to understand the concepts:


```python
// Create: src/LearningComponent.js
import React, { useState } from 'react';
import Plot from 'react-plotly.js';

function LearningComponent() {
  // State is like a variable that triggers re-rendering
  const [threshold, setThreshold] = useState(100);
  
  // Mock data (like your mouse data)
  const mockMice = {
    'Control': [45, 52, 48, 51, 49],
    'Treatment': [65, 72, 68, 71, 69]
  };
  
  // Apply threshold effect (simulate your analysis)
  const applyThreshold = (values, threshold) => {
    return values.map(val => val * (threshold / 255));
  };
  
  return (
    <div style={{ padding: '20px' }}>
      <h2>ðŸ”¬ Learning: How Thresholds Work</h2>
      
      {/* This is like a Jupyter slider widget */}
      <div>
        <label>Threshold: {threshold}</label>
        <input 
          type="range" 
          min="0" 
          max="255" 
          value={threshold}
          onChange={(e) => setThreshold(Number(e.target.value))}
        />
      </div>
      
      {/* This creates a plot like matplotlib */}
      <Plot
        data={[
          {
            y: applyThreshold(mockMice.Control, threshold),
            type: 'box',
            name: 'Control',
            boxpoints: 'all'
          },
          {
            y: applyThreshold(mockMice.Treatment, threshold),
            type: 'box',
            name: 'Treatment',
            boxpoints: 'all'
          }
        ]}
        layout={{
          title: `Effect of Threshold: ${threshold}`,
          yaxis: { title: 'Processed Values' }
        }}
      />
      
      <div style={{ marginTop: '20px', fontSize: '14px' }}>
        <p><strong>ðŸŽ¯ What's happening:</strong></p>
        <ul>
          <li>Moving the slider changes the `threshold` state</li>
          <li>React automatically re-runs the component</li>
          <li>New data is calculated and plot updates</li>
          <li>This is exactly how your real app works!</li>
        </ul>
      </div>
    </div>
  );
}

export default LearningComponent;
```
